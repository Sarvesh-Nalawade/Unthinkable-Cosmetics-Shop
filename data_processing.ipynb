#### Rm Links:
- There are some prod links in `Product Id` col
- or else, not all entries are 32 len hashes
- So replacing all of them with MD5 based 32 len hash
- Doing this as in any ways prod id does not matter (by meaning or name)

```python
len('b77f3da33be6e65f6183da6ada8c07ca')
```

```python
import hashlib
def hash_md5(s: str) -> str:
    """Return MD5 hex digest (32 hex chars)."""
    return hashlib.md5(s.encode('utf-8')).hexdigest()

hash_md5("Hello")
```

```python
# use lambda to hash all the cols where len of the prod id is not 32, else just str(x)
# df['Product Id'] = df['Product Id'].apply(lambda x: hash_md5(x) if len(x) != 32 else str(x))
# df['Product Id'].str.len().value_counts()
```

```python
# Here, we can either hash just the prod ids, or we can also has names with it
# but, realized that prod ids are duplicate already, and since original ids were wrong and duplicated
# hashing just ids produced multiple same hashes, for prods with same id but diff names, 
# hence, decided to hash ids+names both
# Mind u that, even after there, there are still some collision as the df contains some duplicate entries, but listed under diff categories.
# They will later (if) will be dealt with separately

df['Product Id'] = df.apply(
    lambda row: (
        # hash_md5(str(row['Product Id']))
        hash_md5(str(row['Product Name']))
        # hash_md5(str(row['Product Id']) + str(row['Product Name']))
        if len(str(row['Product Id'])) != 32
        else str(row['Product Id'])
    ),
    axis=1
)
```

### Use proper formats:

```python
# 1. Convert Product Price to float
df['Product Price'] = df['Product Price'].astype(float)

# 2. Convert Product Rating to float (some might be missing, so use pd.to_numeric)
df['Product Rating'] = pd.to_numeric(df['Product Rating'], errors='coerce')

# 3. Convert Product Reviews Count to float
df['Product Reviews Count'] = pd.to_numeric(df['Product Reviews Count'], errors='coerce')

# 4. Convert Expected Category Count to int (or float if some missing)
df['Expected Category Count'] = pd.to_numeric(
    df['Expected Category Count'], errors='coerce').astype('Int64')  # nullable integer

# 5. Convert Expected Brand Count to int
df['Expected Brand Count'] = pd.to_numeric(
    df['Expected Brand Count'], errors='coerce').astype('Int64')  # nullable integer
```

```python
df.info()
```

### Fill NAs:

```python
# Fill missing Product Rating with mean
df['Product Rating'] = df['Product Rating'].fillna(df['Product Rating'].mean())

# Fill missing Product Reviews Count with median, then convert to integer
df['Product Reviews Count'] = df['Product Reviews Count'].fillna(
    df['Product Reviews Count'].median()).astype('Int64')
```

```python
df["Product Rating"] = df["Product Rating"].round(1)
```
