## Fundamental CSV issue:

```python
df_more_than_32 = df[df['Product Id'].str.len() != 32]
df_more_than_32.sample(5)
```

```python
# df_mt32_links = df_more_than_32[~df_more_than_32['Product Id'].str.contains("BZ")]
df_mt32_links = df_more_than_32[df_more_than_32['Product Id'].str.startswith("http")]
print(df_mt32_links.shape)
df_mt32_links.sample(5)
```

```python
col_headers = list(df.columns)
# for all rows in df_mt32_links, keep the prod id as it is, and then ignoring the curr 'product brand code' col entry, pull all the cols left by 1 each, and in last col keep na
# i want this in a list of jsons

edited_link_rows = []
for index, row in df_mt32_links.iterrows():
    new_row = {}
    new_row['Product Id'] = row['Product Id']
    for i in range(1, len(col_headers)-1):
        new_row[col_headers[i]] = row[col_headers[i+1]]
    new_row[col_headers[-1]] = None
    edited_link_rows.append(new_row)

print(edited_link_rows[0])
len(edited_link_rows)
```

```python
df[df['Product Id'].str.startswith("http")].shape
```

```python
# just keep the rows without http starting
df = df[~df['Product Id'].str.startswith("http")]
df.shape
```

```python
df.info()
```

```python
# make df from edited_link_rows and append to df
df_edited_links = pd.DataFrame(edited_link_rows)
df_edited_links.sample(2)
```

```python
# append that part in main df again:
df = pd.concat([df, df_edited_links], ignore_index=True)
df.shape
```

### Rm Links From First Col:
- There are some prod links in the first col, or else, not all entries are 32 len hashes
- So replacing all of them with MD5 based 32 len hash

```python
len('b77f3da33be6e65f6183da6ada8c07ca')
```

```python
import hashlib

def hash_md5(s: str) -> str:
    """Return MD5 hex digest (32 hex chars)."""
    return hashlib.md5(s.encode('utf-8')).hexdigest()
```

```python
hash_md5("Hello")
```

```python
# use lambda to hash all the cols where len of the prod id is not 32, else just str(x)
df['Product Id'] = df['Product Id'].apply(lambda x: hash_md5(x) if len(x) != 32 else str(x))
df['Product Id'].str.len().value_counts()
```
